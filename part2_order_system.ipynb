```python
#Task 2 — Cart Operations (8 marks)
menu = {
    "Paneer Tikka":   {"category": "Starters",  "price": 180.0, "available": True},
    "Chicken Wings":  {"category": "Starters",  "price": 220.0, "available": False},
    "Veg Soup":       {"category": "Starters",  "price": 120.0, "available": True},
    "Butter Chicken": {"category": "Mains",     "price": 320.0, "available": True},
    "Dal Tadka":      {"category": "Mains",     "price": 180.0, "available": True},
    "Veg Biryani":    {"category": "Mains",     "price": 250.0, "available": True},
    "Garlic Naan":    {"category": "Mains",     "price":  40.0, "available": True},
    "Gulab Jamun":    {"category": "Desserts",  "price":  90.0, "available": True},
    "Rasgulla":       {"category": "Desserts",  "price":  80.0, "available": True},
    "Ice Cream":      {"category": "Desserts",  "price": 110.0, "available": False},
}
cart =[]
print(cart)
def update_quantity(quant,new_value):
    quant+=new_value
    return quant
    
def add_item_to_cart(new_order,order_quant):
    if new_order not in menu:
        print("The item is not present in the menu")
        return
    if not menu[new_order]["available"]:
        print("Item is unavailable")
        return
    found = False
    for order in cart:
        if order["item"].strip().lower() == new_order.strip().lower():
            order["quantity"] += order_quant
            found = True
            break
    if not found:
           cart.append({
            "item": new_order,
            "quantity": order_quant,
            "price": menu[new_order]["price"]})
            

#Write logic to remove an item from the cart by name.
def remove_item(to_remove):
    found = False
    for index,orders in enumerate(cart):
        if orders["item"]==to_remove:
            found=True
            if orders["quantity"]>=1:
                orders["quantity"]-=1
                if orders["quantity"]==0: 
                    cart.pop(index)
            else:
                cart.pop(index)
            break
    if not found:
        print("The given item doesn't exist in the cart")
    
    
add_item_to_cart("Paneer Tikka",2)
print(cart)
add_item_to_cart("Gulab Jamun",1)
print(cart)
add_item_to_cart("Paneer Tikka",1)
print(cart)
add_item_to_cart("Mystery Burger",1)
print(cart)
add_item_to_cart("Chicken Wings",1)
print(cart)
remove_item("Gulab Jamun")
print(cart)

```

```python
#5.Order Summary
print("\n========== Order Summary ==========")

subtotal = 0

for item in cart:
    total_price = item["quantity"] * item["price"]
    subtotal += total_price
    print(f"{item['item']:20} x{item['quantity']}    ₹{total_price:.2f}")

print("------------------------------------")

gst = subtotal * 0.05
total = subtotal + gst

print(f"Subtotal:                ₹{subtotal:.2f}")
print(f"GST (5%):               ₹{gst:.2f}")
print(f"Total Payable:          ₹{total:.2f}")
print("====================================")
```

```python
#Task 4 — Daily Sales Log Analysis 
sales_log = {
    "2025-01-01": [
        {"order_id": 1,  "items": ["Paneer Tikka", "Garlic Naan"],          "total": 220.0},
        {"order_id": 2,  "items": ["Gulab Jamun", "Veg Soup"],              "total": 210.0},
        {"order_id": 3,  "items": ["Butter Chicken", "Garlic Naan"],        "total": 360.0},
    ],
    "2025-01-02": [
        {"order_id": 4,  "items": ["Dal Tadka", "Garlic Naan"],             "total": 220.0},
        {"order_id": 5,  "items": ["Veg Biryani", "Gulab Jamun"],           "total": 340.0},
    ],
    "2025-01-03": [
        {"order_id": 6,  "items": ["Paneer Tikka", "Rasgulla"],             "total": 260.0},
        {"order_id": 7,  "items": ["Butter Chicken", "Veg Biryani"],        "total": 570.0},
        {"order_id": 8,  "items": ["Garlic Naan", "Gulab Jamun"],           "total": 130.0},
    ],
    "2025-01-04": [
        {"order_id": 9,  "items": ["Dal Tadka", "Garlic Naan", "Rasgulla"], "total": 300.0},
        {"order_id": 10, "items": ["Paneer Tikka", "Gulab Jamun"],          "total": 270.0},
    ],
}
#Print total revenue per day.
highest_selling=float('inf')
daily_revenue = {}
for dates,values in sales_log.items():
    total_revenue=0
    for value in values:
       total_revenue+=value["total"]
    daily_revenue[dates]=total_revenue
    print(f"revenue for {dates} is :{total_revenue}")
print(daily_revenue)
#Print the best-selling day (date with the highest total revenue).
best_day = max(daily_revenue, key=daily_revenue.get)
print(f" Best-Selling Day: {best_day} ({daily_revenue[best_day]})")

#Find the most ordered item
item_count = {}
for orders in sales_log.values():
    for order in orders:
        for item in order["items"]:
            item_count[item] = item_count.get(item, 0) + 1
most_ordered = max(item_count, key=item_count.get)
print(f"Most Ordered Item: {most_ordered} ({item_count[most_ordered]} times)")


```
